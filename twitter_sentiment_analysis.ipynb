```python
import pandas as pd
tweets=pd.read_csv('Tweets.csv')
tweets.head()
```

```python
#We need only actual text and the airline sentiment here i.e. the input and output of the NLP model

#‘airline_sentiment_confidence’ - how sure you are about the sentiment of the tweet. We drop all tweets which have lower confidence than 0.5 (50% sure about the correct classification of sentiment)
#axis=0 --> drop rows when confi<0.5
#shape--> tweet vs parameter number

tweets_df=tweets.drop(tweets[tweets['airline_sentiment_confidence']<0.5].index,axis=0)
tweets_df.shape
```

```python
#Segregate the input and output parameters
X=tweets_df['text']
y=tweets_df['airline_sentiment']
```

```python
# 1. Removing the stop words
# 2. Removing Punctuation
# 3. Stemming - Not all variations of the same word need to be converted in a number

from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
```

```python
#use a portstemmer, reduce the dimentionality of words.
"""prepare our text data for analysis by removing common stopwords and reducing words to their
base form using a PorterStemmer algorithm."""

stop_words=stopwords.words('english')
stemmer=PorterStemmer()
```

```python
''' cleaned_data --> empty list to store text data after preprocessing.
1.Import a library re(regular expressions)
2.So we are basically taking one tweet at a time, and inside it whatever is not
a letter(belonging to a-z or A-Z), will be replaced by an empty space. It will
automatically filter out punctuation marks and other non-letters.
3.we convert all words into lower cases and split them into a list.'''

import re
cleaned_data=[]
for i in range(len(X)):
   tweet=re.sub('[^a-zA-Z]',' ',X.iloc[i])
   tweet=tweet.lower().split()
   t=[stemmer.stem(word) for word in tweet if (word not in stop_words)]
   tweet=' '.join(tweet)
   cleaned_data.append(tweet)
```

```python
#bag of words approach, order of sentence is lost.
#Count Vectorizer function converts a list of words into bag of words
'''Also we set the stop_words parameter to the name of the airline companies,
as we want to remove that from the tweets as well.'''
#cv.fit_transform --> take cleaned_data and convert to bag of words

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000,stop_words=['virginamerica','unit'])
X_fin=cv.fit_transform(cleaned_data).toarray()
```

```python
#replace sentiments into numbers

'''Basically for every word in the y column, we are passing it through a function
 called as ‘sentiment_ordering.index()’ which basically replaces the word with it’s index in the sentiment_ordering list.'''

sentiment_ordering = ['negative', 'neutral', 'positive']
y = y.apply(lambda x: sentiment_ordering.index(x))
```

```python
#Multinomial Naive Bayes model to figure out the relationship between the input and the output
#fit the model

from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
```

```python
'''To build the model we split the dataset into training and tesiting
section --> 30% of the actual data.'''

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_fin,y,test_size=0.3)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
from sklearn.metrics import classification_report
cf=classification_report(y_test,y_pred)
print(cf)
```
